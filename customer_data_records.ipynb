```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
import pandas as pd

# Load the data set
df= pd.read_csv('Customer_data_records.csv')

#Display the dataset
df.head(20)
```

```python
# Select features for clustering
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
#'n_clusters=2' divides the data into 2 clusters (groups).
#'random_state=0'  Sets the seed for random number generation so i get reproducible results every time i run the code
#kmeans.fit(x) Looks for dataset x, Finds 2 cluster centers and Assigns each data point in x to the nearest cluster center
kmeans=KMeans(n_clusters=2,random_state=0)
kmeans.fit(X) 
```

```python
df['cluster']=kmeans.labels_
df
```

```python
plt.figure(figsize=(15,12))
sns.scatterplot(data=df,x='Annual Income (k$)',y='Spending Score (1-100)',hue='cluster')

plt.title('Customer CLusters based on Income&Spending')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
